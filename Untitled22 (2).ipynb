```python
from pymongo import MongoClient
import pandas as pd
import schedule
import time
from datetime import datetime

mongo_url = "mongodb+srv://nirav:nirav@cluster0.wtyu5.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"

mongo_url = mongo_url.replace("<username>", "nirav").replace("<password>", "nirav").replace("<dbname>", "online_retail")

client = MongoClient(mongo_url)
db = client['online_retail']  # Database name
collection = db['products']   # Collection name

def fetch_new_data():
    print(f"Fetching new data at {datetime.now()}...")
    
    
    new_data = pd.DataFrame({
        "Description": ["New Product A", "New Product B"],
        "Quantity": [10, 20],
        "UnitPrice": [5.0, 15.0],
        "Country": ["USA", "Canada"]
    })
    print("New data fetched:")
    print(new_data)

    
    records = new_data.to_dict("records")
    
    
    try:
        result = collection.insert_many(records)
        print(f"Inserted {len(result.inserted_ids)} new records into MongoDB!")
    except Exception as e:
        print(f"Error inserting records: {e}")


schedule.every(24).hours.do(fetch_new_data)


print("Batch process started. Waiting for the next scheduled task...")
while True:
    schedule.run_pending()
    time.sleep(1)
```

```python
from flask import Flask, jsonify, request
from pymongo import MongoClient
from bson.objectid import ObjectId
import threading
from werkzeug.serving import run_simple


mongo_url = "mongodb+srv://nirav:nirav@cluster0.wtyu5.mongodb.net/?retryWrites=true&w=majority&appName=Cluster0"


client = MongoClient(mongo_url)

db = client['online_retail'] 
collection = db['products']   


app = Flask(__name__)

# Route 1: Get all items
@app.route('/api/items', methods=['GET'])
def get_all_items():
    items = list(collection.find({}, {'_id': 1, 'Description': 1, 'Quantity': 1, 'UnitPrice': 1, 'Country': 1}))
    for item in items:
        item['_id'] = str(item['_id'])  
    return jsonify(items)

# Route 2: Get a range of items
@app.route('/api/items/range', methods=['GET'])
def get_items_in_range():
    min_quantity = int(request.args.get('min', 0))
    max_quantity = int(request.args.get('max', 1000))

    items = list(collection.find(
        {'Quantity': {'$gte': min_quantity, '$lte': max_quantity}},
        {'_id': 1, 'Description': 1, 'Quantity': 1, 'UnitPrice': 1, 'Country': 1}
    ))
    for item in items:
        item['_id'] = str(item['_id'])  
    return jsonify(items)

# Route 3: Get item by ID
@app.route('/api/items/<item_id>', methods=['GET'])
def get_item_by_id(item_id):
    try:
        item = collection.find_one({'_id': ObjectId(item_id)})
        if item:
            item['_id'] = str(item['_id'])
            return jsonify(item)
        else:
            return jsonify({'error': 'Item not found'}), 404
    except Exception as e:
        return jsonify({'error': 'Invalid ID format'}), 400

# Route 4: API documentation
@app.route('/', methods=['GET'])
def api_documentation():
    return jsonify({
        "message": "Welcome to the Product API",
        "endpoints": {
            "Get All Items": "/api/items",
            "Get Range of Items": "/api/items/range?min=<min>&max=<max>",
            "Get Item by ID": "/api/items/<item_id>"
        },
        "method": "GET",
        "notes": "Test the API by visiting the provided URLs in your browser or using tools like Postman."
    })


def run_flask():
    run_simple('127.0.0.1', 5000, app, use_reloader=False)


flask_thread = threading.Thread(target=run_flask)
flask_thread.start()
```

```python
from flask import Flask, jsonify, render_template_string
import pandas as pd

# Load data from CSV file
csv_file_path = r"C:\Users\nirav\OneDrive\Desktop\Georgian\DA programming\flask_project\OnlineRetail.csv"  # Replace with your actual file path
try:
    data = pd.read_csv(csv_file_path)
    print("CSV file loaded successfully.")
except FileNotFoundError:
    print(f"Error: CSV file not found at {csv_file_path}.")
    raise
except Exception as e:
    print(f"Error loading CSV file: {e}")
    raise

# Flask app setup
app = Flask(__name__)

# Route to provide JSON data for Google Charts
@app.route('/api/data')
def get_data():
    # Convert CSV data to a list of dictionaries
    json_data = data.to_dict(orient='records')
    return jsonify(json_data)

# HTML template for the dashboard
html_template = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Dashboard</title>
    <script type="text/javascript" src="https://www.gstatic.com/charts/loader.js"></script>
    <script type="text/javascript">
        // Load the Google Charts library
        google.charts.load('current', { packages: ['corechart'] });
        google.charts.setOnLoadCallback(drawChart);

        function drawChart() {
            // Fetch data from Flask API
            fetch('/api/data')
                .then(response => response.json())
                .then(data => {
                    // Prepare the data for Google Charts
                    const chartData = [['Description', 'Quantity']];
                    data.forEach(item => chartData.push([item.Description, item.Quantity]));

                    // Create the data table
                    const dataTable = google.visualization.arrayToDataTable(chartData);

                    // Chart options
                    const options = {
                        title: 'Product Quantities',
                        width: 800,
                        height: 600,
                        hAxis: { title: 'Product Names' },
                        vAxis: { title: 'Quantity' },
                        bar: { groupWidth: '75%' },
                        legend: { position: 'none' }
                    };

                    // Draw the bar chart
                    const chart = new google.visualization.BarChart(document.getElementById('chart_div'));
                    chart.draw(dataTable, options);
                })
                .catch(error => console.error('Error fetching data:', error));
        }
    </script>
</head>
<body>
    <div style="text-align: center;">
        <h1>Product Dashboard</h1>
        <div id="chart_div"></div>
    </div>
</body>
</html>
"""

# Route to render the dashboard
@app.route('/')
def dashboard():
    return render_template_string(html_template)

# Run the Flask app
if __name__ == '__main__':
    app.run(debug=True)
```
